```python
import pandas as pd
import numpy as np
import cv2
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# slide_id: 0ADD705
# frame_id: 236
# cell_x: 1056
# cell_y: 325

df_if = pd.read_table('0ADD705/if_data/frame_236.txt')
df_imc = pd.read_table('0ADD705/imc_data/0ADD705_s0_p1_r43_a43_ac_236_960_full.csv', sep=',')
img_if = cv2.imread('0ADD705/if_img/Tile000236.tif', -1)
img_imc = cv2.imread('0ADD705/imc_img/0ADD705_s0_p1_r43_a43_ac_236_960_full.tiff', -1)

# print(img_if.shape)
# plt.figure(figsize=(18,15))
# plt.imshow(img_if)
# sns.scatterplot(df_if, x='x', y='y', size=1)
# plt.show()

# print(img_imc.shape)
# plt.figure(figsize=(12,12))
# plt.imshow(np.transpose(img_imc))
# sns.scatterplot(df_imc, x='x', y='y', size=0.5)
# plt.show()

if_origin = [1056, 325]
imc_origin = [200, 200]
# center the origin and IF scaling
df_if.loc[:, ['x', 'y']] = 0.59*(df_if.loc[:, ['x', 'y']] - if_origin) 
df_imc.loc[:, ['x', 'y']] = df_imc.loc[:, ['x', 'y']] - imc_origin

df_imc_temp = pd.DataFrame()
# flipping : x->y, y->x
df_imc_temp['x'] = df_imc.loc[:, 'y']
df_imc_temp['y'] = df_imc.loc[:, 'x']
df_imc = df_imc_temp.copy()

df_if['mode'] = 'IF'
df_imc['mode'] = 'IMC'
df = pd.concat([df_if.loc[:, ['x','y','mode']], df_imc.loc[:, ['x','y','mode']]])

plt.figure(figsize=(18,18))
sns.scatterplot(df, x='x', y='y', hue='mode')
plt.show()
```

# Filtering all IMC ROIs

```python
import glob
imc_paths = glob.glob('data/0B11619/imc/feature/*')
for path in imc_paths:
    df_imc = pd.read_table(path)
    name = path.split('/')[-1]
    print(name)
    coord_filter = (df_imc.x < 300) & (df_imc.x > 100) & (df_imc.y < 310) & (df_imc.y > 90) 
    df_imc = df_imc[coord_filter]
    
    
    df_imc.to_csv(f'data/0B11619/mappingIMC/{name}', index=False, sep='\t')
```
